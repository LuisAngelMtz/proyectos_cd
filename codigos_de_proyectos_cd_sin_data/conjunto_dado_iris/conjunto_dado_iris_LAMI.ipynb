```python
import pandas as pd
import numpy as np

from sklearn import datasets
```

https://pt.wikipedia.org/wiki/Conjunto_de_dados_flor_Iris

# Lectura de datos

```python
iris_file = datasets.load_iris()
```

```python
iris_file
```

Descripción del problema:

- Este conjunto de datos consta de 3 tipos diferentes de iris (Setosa, Versicolor, Virginica), con las longitudes de pétalos y sépalos. ( array de 150x4 + tgt)

- Las filas son las muestras de las plantas y las columnas son: longitud del sépalo, anchura del sépalo, longitud del pétalo y anchura del pétalo

# Tratamiento / Limpieza

```python
df_iris = pd.DataFrame( data = np.c_[iris_file['data'], iris_file['target']] ,
                        columns = iris_file['feature_names'] + ['clase'] )
```

```python
df_iris
```

```python
cols =  df_iris.columns.to_list()
```

```python
cols = [c for c in cols if c not in ['clase']]
```

# Estandarización

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
```

```python
ss = StandardScaler()
```

```python
df_iris[ ['ss_'+c for c in cols]  ] =  ss.fit_transform( df_iris[cols] )
```

```python
df_iris
```

# Reducción de dimensiones - PCA

```python
cols_ss = df_iris.filter(like='ss_').columns.tolist()
```

## Componentes principales (2)

```python
pca = PCA( n_components=2 )
```

```python
array_pca = pca.fit_transform(  df_iris[cols_ss]  )
```

```python
df_pca = pd.DataFrame({ 'PCA1': array_pca[:,0],
                         'PCA2': array_pca[:,1],
                         'clase': df_iris['clase']})
```

```python
df_pca 
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

### Gráficas Simples

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot( x = ['PCA1','PCA2'], y=pca.explained_variance_ratio_ )
```

```python
sns.scatterplot( x='PCA1',
                 y='PCA2',
                 hue='clase',
                 data=df_pca)
```

## Componentes principales (3)

```python
pca = PCA( n_components=3 )
```

```python
array_pca = pca.fit_transform(  df_iris[cols_ss]  )
```

```python
df_pca = pd.DataFrame({ 'PCA1': array_pca[:,0],
                         'PCA2': array_pca[:,1],
                         'PCA3': array_pca[:,2],
                         'clase': df_iris['clase']})
```

```python
df_pca  # esta sería la TAD 
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

### Gráficas simples

```python
sns.barplot( x = ['PCA1','PCA2','PCA3'], y=pca.explained_variance_ratio_ )
```

```python
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(111,projection='3d')

x = df_pca['PCA1']
y = df_pca['PCA2']
z = df_pca['PCA3']

col = df_pca['clase'].map( {0:'r', 1:'b', 2:'g'} )

ax.scatter( x,y,z , c=col, marker='o')

ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')

plt.show()
```
